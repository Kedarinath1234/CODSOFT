```python
# I created a class named Task

class Task:
    def __init__(self, description, completed=False):
        self.description = description
        self.completed = completed
        
# Here I am creating status of the task that it is completed or not 

    def __str__(self):
        status = "✓" if self.completed else " "
        return f"[{status}] {self.description}"
    
# Here I am creating a class of ToDoList and passing self in it 
# In this class we can easily add a New Task of our choice

class Todolist:
    def __init__(self):
        self.tasks = []
        
# Here I am adding for the new Task to Take next input from User 
# It will add the new Task to the To Do List

    def add_task(self, description):
        task = Task(description)
        self.tasks.append(task)
        print(f"Task '{description}' added to the To-Do List.")
        
# Here I can Check out all my Tasks which I have been added to the To Do list
# Here I am initializing Index = 1 as Default
# Here for loop is used for iteration 
    def view_tasks(self):
        if not self.tasks:
            print("Your To-Do List has no Task.")
        else:
            print("To-Do List:")
            for index, task in enumerate(self.tasks, 1):
                print(f"{index}. {task}")
                
# Here I confirming that my task has been completed 
# Here If Index number is not found then it will not print Invalid Task number

    def complete_task(self, task_index):
        if 1 <= task_index <= len(self.tasks):
            task = self.tasks[task_index - 1]
            task.completed = True
            print(f"Task '{task.description}' marked as complete (✓).")
        else:
            print("Invalid task index Number. Please enter the valid task number.")
            
# Here I combines all the Task Specfications Using While and If- else conditions
# Here By main function I combined all the Task Specifications and created a object to call them with ease

def main():
    todo_list = Todolist()

    while True:
        print("\nMenu:")
        print("1. Add New Task")
        print("2. View Tasks")
        print("3. Mark Task as Complete")
        print("4. Quit")

        choice = input("Enter your choice: ")

        if choice == '1':
            task = input("Enter the task that you want complete: ")
            todo_list.add_task(task)
        elif choice == '2':
            todo_list.view_tasks()
        elif choice == '3':
            task_index = int(input("Enter the task number to mark as complete (✓): "))
            todo_list.complete_task(task_index)
        elif choice == '4':
            print("Goodbye and have a great Day !!!")
            break
        else:
            print("Invalid choice. Please select a valid option from the given menu.")

if __name__ == "__main__":
    main()
```
