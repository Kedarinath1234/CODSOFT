```python
# using while for iteration
while True:
    print('\nCalculator')
    print('1.Addition')
    print('2.Substraction')
    print('3.Multiply')
    print('4.Division')
    print('5.Exit')

# creating choice as input
    choice=int(input('Enter you choice between (1 - 5): '))

# if-elif-else for multiple c
    if choice == 1:
        A = int(input('Enter the value of A : '))
        B = int(input('Enter the value of B : '))
        Add = A + B
        print('Addition of two Numbers = ',Add)
    
    elif choice == 2:
        C = int(input('Enter the Value of C : '))
        D = int(input('Enter the value of D : '))
        Sub = C - D
        print('Substraction of two Numbers = ',Sub)
        
    elif choice == 3:
        E = int(input('Enter the value of E : '))
        F = int(input('Enter the value of F : '))
        Mul = E * F
        print('Multiplication of two Numbers = ',Mul)
            
    elif choice == 4:
        G = int(input('Enter the value of G : '))
        H = int(input('Enter the value of H : '))
        Div = G / H
        print('Division of two Numbers = ',Div)
    
    elif choice == 5:
        print('Exiting the Calculator.GoodBye !')
        break
    else:
        print('choice is Invalid.Please Enter the Choice (1 - 5)')
```
